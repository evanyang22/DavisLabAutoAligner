```python
import pandas as pd
import numpy as np
from scipy import stats
data = pd.read_csv("JuJu_EV_counts.csv")
ColumnHeaders=list(data.columns)
#masterDF=pd.DataFrame

i=1
while i< len(ColumnHeaders):
    if i==1 or i%3==1:
        NewColumnName= ColumnHeaders[i][:-2]
        newDF=data[["Genes",NewColumnName+"_1",NewColumnName+"_2",NewColumnName+"_3"]]
        #finding GM
        newDF[NewColumnName+'_GM'] = None
        newDF[NewColumnName+"_GM"] = stats.gmean(newDF.iloc[:, 1:4], axis=1)
        
        if i==1:
            masterDF=newDF
        if i!=1:
            masterDF=masterDF.merge(newDF)
        
        
    i+=1
```

```python
masterDF.to_csv("JuJu_EV_counts_GM.csv",index=False)
```
